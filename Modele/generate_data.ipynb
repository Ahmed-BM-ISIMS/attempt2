```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Define diseases with prevalence, duration, age ranges, pain levels, male_bias, typical comorbidities, typical symptom duration (in hours), temperature ranges, etc.
diseases = [
    {'name': 'Heart Attack', 'prevalence': 0.05, 'duration': (90, 150), 'age_range': (45, 80), 'pain_levels': {'Severe': 0.7, 'Moderate': 0.2, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.7, 'comorbidities': {'Diabetes': 0.4, 'Hypertension': 0.6, 'Previous Heart Disease': 0.5, 'None': 0.3}, 'symptom_duration_hours': (0.5, 6), 'temp_range': (36.5, 37.5), 'rr_range_normal': (12, 20), 'rr_range_urgent': (20, 30), 'spo2_range_normal': (95, 100), 'spo2_range_low': (85, 95)},
    {'name': 'Appendicitis', 'prevalence': 0.10, 'duration': (45, 75), 'age_range': (18, 60), 'pain_levels': {'Severe': 0.6, 'Moderate': 0.3, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.5, 'comorbidities': {'None': 0.8, 'Obesity': 0.2}, 'symptom_duration_hours': (12, 48), 'temp_range': (37.5, 39.0), 'rr_range_normal': (12, 20), 'rr_range_urgent': (18, 25), 'spo2_range_normal': (95, 100), 'spo2_range_low': (92, 98)},
    {'name': 'Gallstones', 'prevalence': 0.14, 'duration': (40, 60), 'age_range': (30, 70), 'pain_levels': {'Severe': 0.4, 'Moderate': 0.4, 'Mild': 0.2, 'None': 0.0}, 'male_bias': 0.4, 'comorbidities': {'Obesity': 0.5, 'Diabetes': 0.3, 'None': 0.4}, 'symptom_duration_hours': (2, 24), 'temp_range': (36.5, 38.0), 'rr_range_normal': (12, 20), 'rr_range_urgent': (15, 25), 'spo2_range_normal': (95, 100), 'spo2_range_low': (92, 98)},
    {'name': 'Hernia', 'prevalence': 0.14, 'duration': (50, 80), 'age_range': (25, 70), 'pain_levels': {'Severe': 0.3, 'Moderate': 0.5, 'Mild': 0.2, 'None': 0.0}, 'male_bias': 0.6, 'comorbidities': {'Obesity': 0.4, 'Previous Surgery': 0.3, 'None': 0.5}, 'symptom_duration_hours': (24, 168), 'temp_range': (36.5, 37.5), 'rr_range_normal': (12, 20), 'rr_range_urgent': (15, 25), 'spo2_range_normal': (95, 100), 'spo2_range_low': (90, 95)},
    {'name': 'Dislocated Knee', 'prevalence': 0.08, 'duration': (30, 60), 'age_range': (18, 50), 'pain_levels': {'Severe': 0.5, 'Moderate': 0.4, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.5, 'comorbidities': {'Previous Knee Injury': 0.3, 'None': 0.7}, 'symptom_duration_hours': (0.1, 2), 'temp_range': (36.5, 37.5), 'rr_range_normal': (12, 20), 'rr_range_urgent': (18, 28), 'spo2_range_normal': (95, 100), 'spo2_range_low': (92, 98)},
    {'name': 'Fracture', 'prevalence': 0.16, 'duration': (30, 60), 'age_range': (18, 80), 'pain_levels': {'Severe': 0.3, 'Moderate': 0.4, 'Mild': 0.2, 'None': 0.1}, 'male_bias': 0.5, 'comorbidities': {'Osteoporosis': 0.2, 'None': 0.8}, 'symptom_duration_hours': (0.1, 2), 'temp_range': (36.5, 37.5), 'rr_range_normal': (12, 20), 'rr_range_urgent': (18, 28), 'spo2_range_normal': (95, 100), 'spo2_range_low': (92, 98)},
    {'name': 'Stroke', 'prevalence': 0.05, 'duration': (90, 180), 'age_range': (50, 80), 'pain_levels': {'Severe': 0.1, 'Moderate': 0.2, 'Mild': 0.4, 'None': 0.3}, 'male_bias': 0.55, 'comorbidities': {'Hypertension': 0.7, 'Diabetes': 0.4, 'Heart Disease': 0.3, 'None': 0.2}, 'symptom_duration_hours': (0.5, 6), 'temp_range': (36.5, 37.5), 'rr_range_normal': (12, 20), 'rr_range_urgent': (20, 30), 'spo2_range_normal': (95, 100), 'spo2_range_low': (80, 95)},
    {'name': 'Kidney Stones', 'prevalence': 0.12, 'duration': (30, 60), 'age_range': (30, 60), 'pain_levels': {'Severe': 0.5, 'Moderate': 0.4, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.6, 'comorbidities': {'Obesity': 0.3, 'Hypertension': 0.2, 'None': 0.6}, 'symptom_duration_hours': (2, 48), 'temp_range': (36.5, 37.5), 'rr_range_normal': (12, 20), 'rr_range_urgent': (18, 28), 'spo2_range_normal': (95, 100), 'spo2_range_low': (92, 98)},
    {'name': 'Torn ACL', 'prevalence': 0.08, 'duration': (60, 90), 'age_range': (18, 40), 'pain_levels': {'Severe': 0.4, 'Moderate': 0.5, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.5, 'comorbidities': {'Previous Knee Injury': 0.4, 'None': 0.6}, 'symptom_duration_hours': (0.1, 2), 'temp_range': (36.5, 37.5), 'rr_range_normal': (12, 20), 'rr_range_urgent': (18, 28), 'spo2_range_normal': (95, 100), 'spo2_range_low': (92, 98)},
    {'name': 'Bowel Obstruction', 'prevalence': 0.08, 'duration': (60, 120), 'age_range': (40, 80), 'pain_levels': {'Severe': 0.5, 'Moderate': 0.3, 'Mild': 0.2, 'None': 0.0}, 'male_bias': 0.5, 'comorbidities': {'Previous Surgery': 0.5, 'Crohn\'s': 0.2, 'None': 0.4}, 'symptom_duration_hours': (6, 48), 'temp_range': (36.5, 38.5), 'rr_range_normal': (12, 20), 'rr_range_urgent': (20, 30), 'spo2_range_normal': (95, 100), 'spo2_range_low': (85, 95)}
]

# Allergies options with probabilities (common in surgical patients)
allergies_options = {'None': 0.8, 'Penicillin': 0.08, 'Cephalosporins': 0.04, 'Antibiotics (Other)': 0.04, 'Latex': 0.02, 'Other': 0.02}

# Verify prevalence sums to 1
total_prev = sum(d['prevalence'] for d in diseases)
if abs(total_prev - 1.0) > 0.01:
    raise ValueError(f"Prevalence must sum to 1, got {total_prev}")

# Generate 1,000 patients
data = []
for i in range(1000):
    # Select disease based on prevalence
    condition_info = np.random.choice(diseases, p=[d['prevalence'] for d in diseases])
    condition = condition_info['name']
    
    # Gender based on condition-specific bias
    gender = np.random.choice(['Male', 'Female'], p=[condition_info['male_bias'], 1 - condition_info['male_bias']])
    
    # Age skewed by condition
    age = np.random.randint(condition_info['age_range'][0], condition_info['age_range'][1])
    
    # Pain level based on condition
    pain_level = np.random.choice(
        list(condition_info['pain_levels'].keys()),
        p=list(condition_info['pain_levels'].values())
    )
    
    # Heart rate and blood pressure influenced by condition and pain
    if condition in ['Heart Attack', 'Appendicitis', 'Stroke', 'Bowel Obstruction'] or pain_level == 'Severe':
        heart_rate = np.random.randint(80, 130)  # Higher for urgent/painful conditions
        systolic = np.random.randint(110, 150)
        diastolic = np.random.randint(70, 100)
        respiratory_rate = np.random.randint(condition_info['rr_range_urgent'][0], condition_info['rr_range_urgent'][1])
        oxygen_saturation = np.random.randint(condition_info['spo2_range_low'][0], condition_info['spo2_range_low'][1])
        temperature = np.random.uniform(condition_info['temp_range'][0] + 0.5, condition_info['temp_range'][1] + 0.5) if 'fever' in condition.lower() else np.random.uniform(condition_info['temp_range'][0], condition_info['temp_range'][1])
    else:
        heart_rate = np.random.randint(60, 100)  # Normal range
        systolic = np.random.randint(90, 130)
        diastolic = np.random.randint(60, 85)
        respiratory_rate = np.random.randint(condition_info['rr_range_normal'][0], condition_info['rr_range_normal'][1])
        oxygen_saturation = np.random.randint(condition_info['spo2_range_normal'][0], condition_info['spo2_range_normal'][1])
        temperature = np.random.uniform(condition_info['temp_range'][0], condition_info['temp_range'][1])
    
    # Surgery duration
    duration = np.random.randint(condition_info['duration'][0], condition_info['duration'][1])
    
    # Allergies
    allergy = np.random.choice(list(allergies_options.keys()), p=list(allergies_options.values()))
    
    # Known Medical History (select one or none)
    medical_history = np.random.choice(list(condition_info['comorbidities'].keys()), p=list(condition_info['comorbidities'].values()))
    
    # Symptom Duration (in hours)
    symptom_duration_hours = np.random.uniform(condition_info['symptom_duration_hours'][0], condition_info['symptom_duration_hours'][1])
    
    # Reason for Visit (simple description based on condition)
    reason_for_visit = f"Acute pain due to {condition.lower()}" if symptom_duration_hours < 24 else f"Persistent symptoms from {condition.lower()}"

    data.append({
        'patient_id': f'P{i+1:03d}',
        'gender': gender,
        'age': age,
        'heart_rate': heart_rate,
        'respiratory_rate': respiratory_rate,
        'oxygen_saturation': oxygen_saturation,
        'temperature': round(temperature, 1),
        'blood_pressure': f'{systolic}/{diastolic}',
        'condition': condition,
        'pain_level': pain_level,
        'allergy': allergy,
        'medical_history': medical_history,
        'symptom_duration_hours': round(symptom_duration_hours, 1),
        'reason_for_visit': reason_for_visit,
        'surgery_duration': duration
    })

# Create DataFrame and save to CSV
df = pd.DataFrame(data)
df.to_csv('patients_advanced.csv', index=False)
print("Generated patients_advanced.csv with 1000 rows")
print(df.head())
print("\nDisease Distribution:")
print(df['condition'].value_counts(normalize=True))
print("\nGender Distribution by Condition:")
print(df.groupby('condition')['gender'].value_counts(normalize=True))
```
