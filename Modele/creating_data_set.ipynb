```python
# Add priority column (1-10) based on urgency
def calculate_priority(row):
    # Base priority: Inverse condition code (Heart Attack=1 -> 10, Hernia=10 -> 1)
    base_priority = 11 - row['condition']
    
    # If no surgery needed, low priority (1 or 2)
    if not row['needs_surgery']:
        return min(base_priority, 2)
    
    # Adjustments for urgency
    priority = base_priority
    if row['pain_level'] == 3:  # Severe
        priority += 2
    elif row['pain_level'] == 2:  # Moderate
        priority += 1
    if row['bp_mean'] > 120 or row['bp_mean'] < 80:  # Abnormal BP
        priority += 1
    if row['oxygen_saturation'] < 90:  # Low oxygen
        priority += 1
    if row['age'] > 60 and row['condition'] <= 2:  # Elderly with critical condition
        priority += 1
    
    # Cap priority at 1-10
    return min(max(priority, 1), 10)

# Apply to dataset
df['priority'] = df.apply(calculate_priority, axis=1)

# Verify priority distribution
print("Priority Distribution:")
print(df['priority'].value_counts().sort_index())

# Save updated dataset
df.to_csv('patients_pro_max_with_priority.csv', index=False)
print("Dataset saved as 'patients_pro_max_with_priority.csv'")

# Display sample
print("\nSample with Priority:")
print(df[['patient_id', 'condition', 'needs_surgery', 'pain_level', 'bp_mean', 'oxygen_saturation', 'age', 'priority']].head())
```

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Define diseases with prevalence, duration, age ranges, pain levels, and surgery probability
diseases = [
    {'name': 'Heart Attack', 'prevalence': 0.05, 'duration': (90, 150), 'age_range': (45, 80), 'pain_levels': {'Severe': 0.7, 'Moderate': 0.2, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.7, 'rr_normal': (12, 20), 'rr_urgent': (20, 30), 'spo2_normal': (95, 100), 'spo2_low': (85, 95), 'surgery_probability': 0.3},
    {'name': 'Appendicitis', 'prevalence': 0.10, 'duration': (45, 75), 'age_range': (18, 60), 'pain_levels': {'Severe': 0.6, 'Moderate': 0.3, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.5, 'rr_normal': (12, 20), 'rr_urgent': (18, 25), 'spo2_normal': (95, 100), 'spo2_low': (92, 98), 'surgery_probability': 0.9},
    {'name': 'Gallstones', 'prevalence': 0.14, 'duration': (40, 60), 'age_range': (30, 70), 'pain_levels': {'Severe': 0.4, 'Moderate': 0.4, 'Mild': 0.2, 'None': 0.0}, 'male_bias': 0.4, 'rr_normal': (12, 20), 'rr_urgent': (15, 25), 'spo2_normal': (95, 100), 'spo2_low': (92, 98), 'surgery_probability': 0.8},
    {'name': 'Hernia', 'prevalence': 0.14, 'duration': (50, 80), 'age_range': (25, 70), 'pain_levels': {'Severe': 0.3, 'Moderate': 0.5, 'Mild': 0.2, 'None': 0.0}, 'male_bias': 0.6, 'rr_normal': (12, 20), 'rr_urgent': (15, 25), 'spo2_normal': (95, 100), 'spo2_low': (90, 95), 'surgery_probability': 0.95},
    {'name': 'Dislocated Knee', 'prevalence': 0.08, 'duration': (30, 60), 'age_range': (18, 50), 'pain_levels': {'Severe': 0.5, 'Moderate': 0.4, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.5, 'rr_normal': (12, 20), 'rr_urgent': (18, 28), 'spo2_normal': (95, 100), 'spo2_low': (92, 98), 'surgery_probability': 0.9},
    {'name': 'Fracture', 'prevalence': 0.16, 'duration': (30, 60), 'age_range': (18, 80), 'pain_levels': {'Severe': 0.3, 'Moderate': 0.4, 'Mild': 0.2, 'None': 0.1}, 'male_bias': 0.5, 'rr_normal': (12, 20), 'rr_urgent': (18, 28), 'spo2_normal': (95, 100), 'spo2_low': (92, 98), 'surgery_probability': 0.85},
    {'name': 'Stroke', 'prevalence': 0.05, 'duration': (90, 180), 'age_range': (50, 80), 'pain_levels': {'Severe': 0.1, 'Moderate': 0.2, 'Mild': 0.4, 'None': 0.3}, 'male_bias': 0.55, 'rr_normal': (12, 20), 'rr_urgent': (20, 30), 'spo2_normal': (95, 100), 'spo2_low': (80, 95), 'surgery_probability': 0.2},
    {'name': 'Kidney Stones', 'prevalence': 0.12, 'duration': (30, 60), 'age_range': (30, 60), 'pain_levels': {'Severe': 0.5, 'Moderate': 0.4, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.6, 'rr_normal': (12, 20), 'rr_urgent': (18, 28), 'spo2_normal': (95, 100), 'spo2_low': (92, 98), 'surgery_probability': 0.5},
    {'name': 'Torn ACL', 'prevalence': 0.08, 'duration': (60, 90), 'age_range': (18, 40), 'pain_levels': {'Severe': 0.4, 'Moderate': 0.5, 'Mild': 0.1, 'None': 0.0}, 'male_bias': 0.5, 'rr_normal': (12, 20), 'rr_urgent': (18, 28), 'spo2_normal': (95, 100), 'spo2_low': (92, 98), 'surgery_probability': 0.95},
    {'name': 'Bowel Obstruction', 'prevalence': 0.08, 'duration': (60, 120), 'age_range': (40, 80), 'pain_levels': {'Severe': 0.5, 'Moderate': 0.3, 'Mild': 0.2, 'None': 0.0}, 'male_bias': 0.5, 'rr_normal': (12, 20), 'rr_urgent': (20, 30), 'spo2_normal': (95, 100), 'spo2_low': (85, 95), 'surgery_probability': 0.9}
]

# Verify prevalence sums to 1
total_prev = sum(d['prevalence'] for d in diseases)
if abs(total_prev - 1.0) > 0.01:
    raise ValueError(f"Prevalence must sum to 1, got {total_prev}")

# Generate 15,000 patients
data = []
for i in range(15000):
    disease = np.random.choice(diseases, p=[d['prevalence'] for d in diseases])
    condition = disease['name']
    gender = np.random.choice(['Male', 'Female'], p=[disease['male_bias'], 1 - disease['male_bias']])
    age = np.random.randint(disease['age_range'][0], disease['age_range'][1])
    pain_level = np.random.choice(
        list(disease['pain_levels'].keys()),
        p=list(disease['pain_levels'].values())
    )
    urgent_condition = (pain_level == 'Severe') or (condition in ['Heart Attack', 'Appendicitis', 'Stroke', 'Bowel Obstruction'])
    if urgent_condition:
        heart_rate = np.random.randint(80, 130)
        respiratory_rate = np.random.randint(disease['rr_urgent'][0], disease['rr_urgent'][1])
        oxygen_saturation = np.random.randint(disease['spo2_low'][0], disease['spo2_low'][1])
        systolic = np.random.randint(110, 150)
        diastolic = np.random.randint(70, 100)
    else:
        heart_rate = np.random.randint(60, 100)
        respiratory_rate = np.random.randint(disease['rr_normal'][0], disease['rr_normal'][1])
        oxygen_saturation = np.random.randint(disease['spo2_normal'][0], disease['spo2_normal'][1])
        systolic = np.random.randint(90, 130)
        diastolic = np.random.randint(60, 85)
    needs_surgery = np.random.choice([True, False], p=[disease['surgery_probability'], 1 - disease['surgery_probability']])
    duration = np.random.randint(disease['duration'][0], disease['duration'][1]) if needs_surgery else 0

    data.append({
        'patient_id': f'P{i+1:03d}',
        'age': age,
        'gender': gender,
        'heart_rate': heart_rate,
        'respiratory_rate': respiratory_rate,
        'oxygen_saturation': oxygen_saturation,
        'blood_pressure': f'{systolic}/{diastolic}',
        'pain_level': pain_level,
        'condition': condition,
        'needs_surgery': needs_surgery,
        'surgery_duration': duration
    })

# Create DataFrame
df = pd.DataFrame(data)

# Preprocess for priority calculation
condition_map = {
    'Heart Attack': 1, 'Stroke': 2, 'Dislocated Knee': 3, 'Kidney Stones': 4,
    'Appendicitis': 5, 'Torn ACL': 6, 'Bowel Obstruction': 7, 'Fracture': 8,
    'Gallstones': 9, 'Hernia': 10
}
pain_map = {'None': 0, 'Mild': 1, 'Moderate': 2, 'Severe': 3}
df['condition'] = df['condition'].map(condition_map)
df['pain_level'] = df['pain_level'].map(pain_map).fillna(0)
df['bp_mean'] = df['blood_pressure'].apply(lambda x: (int(x.split('/')[0]) + 2 * int(x.split('/')[1])) / 3)

# Add priority column (1-10) based on urgency
def calculate_priority(row):
    try:
        base_priority = 11 - row['condition']
        if not row['needs_surgery']:
            return min(base_priority, 2)
        priority = base_priority
        if row['pain_level'] == 3:  # Severe
            priority += 2
        elif row['pain_level'] == 2:  # Moderate
            priority += 1
        if row['bp_mean'] > 120 or row['bp_mean'] < 80:  # Abnormal BP
            priority += 1
        if row['oxygen_saturation'] < 90:  # Low oxygen
            priority += 1
        if row['age'] > 60 and row['condition'] <= 2:  # Elderly with critical condition
            priority += 1
        return min(max(priority, 1), 10)
    except (KeyError, TypeError, ValueError):
        print(f"Error processing row: {row}")
        return 1

# Apply to dataset
df['priority'] = df.apply(calculate_priority, axis=1)

# Save updated dataset
df.to_csv('patients_pro_max_with_priority.csv', index=False)
print("Generated patients_pro_max_with_priority.csv with 15000 rows")

# Verify distributions
print("\nDataset Head:")
print(df.head())
print("\nPriority Distribution:")
print(df['priority'].value_counts().sort_index())
print("\nDisease Distribution:")
print(df['condition'].map({v: k for k, v in condition_map.items()}).value_counts(normalize=True))
print("\nGender Distribution by Condition:")
print(df.groupby(df['condition'].map({v: k for k, v in condition_map.items()}))['gender'].value_counts(normalize=True))
print("\nAverage Vitals by Condition:")
print(df.groupby(df['condition'].map({v: k for k, v in condition_map.items()}))[['heart_rate', 'respiratory_rate', 'oxygen_saturation']].mean())
print("\nSurgical Necessity by Condition:")
print(df.groupby(df['condition'].map({v: k for k, v in condition_map.items()}))['needs_surgery'].mean())
print("\nPriority by Condition:")
print(df.groupby(df['condition'].map({v: k for k, v in condition_map.items()}))['priority'].mean())
```
